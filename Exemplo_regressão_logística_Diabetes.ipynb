```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/diabetes.csv')
```

```python
dados.head(3)
```

```python
dados.info()
```

```python
dados.isnull().sum()
```

```python
dados['Outcome'].value_counts()
```

```python
# Separar dados
X = dados.iloc[:, 0:8].values
y = dados.iloc[:, 8].values
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, random_state=7)
```

```python
X_treino.shape, y_treino.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression(max_iter=2000)
```

```python
log.fit(X_treino, y_treino)
```

```python
y_pred = log.predict(X_teste)
```

```python
y_pred == y_teste
```

```python
# Usar o modelo
y_teste[0]
```

```python
log.predict([X_teste[0]])
```

```python
log.coef_
```

```python
log.intercept_
```

```python
log.classes_
```

```python
log.decision_function(X_treino)
```

```python
log.score(X_teste, y_teste)
```
